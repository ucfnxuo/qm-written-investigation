The file you uploaded contains data related to Manchester United's performance in the Premier League for a specific period. Here's a summary of the information in the first few rows:

Date: Dates of the matches (ranging from August 14, 2023, onwards).
Team: Details of the match (e.g., "MAN UTD VS WOLVES").
Win/Lose: Outcome of the match for Manchester United (win or lose).
Score: The score of the match.
Home/Away: Whether the match was played at home or away.
Possession: The possession percentage of Manchester United.
Total Shots: Total number of shots taken.
Shots on Target: Number of shots on target.
Total Passes: Total number of passes made.
Passes Completed: Number of successfully completed passes.
Tackles Won: Number of tackles won.
Aerial Duels Won: Number of aerial duels won.

```python
from sklearn.preprocessing import StandardScaler

# Selecting the numerical columns for standardization
numerical_columns = ['possession', 'total shots', 'shots on target', 'total passes', 'passes completed', 'tackles won', 'aerial duels won']

# Applying standardization to these columns
scaler = StandardScaler()
data_standardized = man_utd_data.copy()
data_standardized[numerical_columns] = scaler.fit_transform(man_utd_data[numerical_columns])

# Displaying the first few rows of the standardized data
data_standardized.head()
```

```python
from datetime import datetime

# Convert 'date' to datetime format and extract the month
man_utd_data['date'] = pd.to_datetime(man_utd_data['date'])
man_utd_data['Month'] = man_utd_data['date'].dt.month_name()

# Define a function to count wins, losses, and draws
def count_results(row):
    if row['win/lose'] == 'win':
        return 'wins'
    elif row['win/lose'] == 'lose':
        return 'losses'
    else:
        return 'draws'

# Apply the function to the dataframe
man_utd_data['result'] = man_utd_data.apply(count_results, axis=1)

# Group by month and calculate the required statistics
man_utd_month = man_utd_data.groupby('Month').agg(
    wins=('result', lambda x: (x == 'wins').sum()),
    losses=('result', lambda x: (x == 'losses').sum()),
    draws=('result', lambda x: (x == 'draws').sum()),
    homes=('home/away', lambda x: (x == 'home').sum()),
    aways=('home/away', lambda x: (x == 'away').sum()),
    scores=('score', 'mean'),
    possession=('possession', 'mean'),
    total_shots=('total shots', 'mean'),
    shots_on_target=('shots on target', 'mean'),
    total_passes=('total passes', 'mean'),
    passes_completed=('passes completed', 'mean'),
    tackles_won=('tackles won', 'mean'),
    aerial_duels_won=('aerial duels won', 'mean')
)

man_utd_month.reset_index(inplace=True)

man_utd_month
```

```python
# Sort the dataframe by month in chronological order
# Since the month names are in full, a custom sorter is needed

months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 
                'August', 'September', 'October', 'November', 'December']
man_utd_month['Month'] = pd.Categorical(man_utd_month['Month'], categories=months_order, ordered=True)

# Sort the dataframe
man_utd_month = man_utd_month.sort_values('Month')

man_utd_month.reset_index(drop=True, inplace=True)
man_utd_month
```

use vif to deal with multicollinearity

```python
# calculating VIF
# This function is adjusted from: https://stackoverflow.com/a/51329496/4667568
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

def drop_column_using_vif_(df, thresh=5):
    '''
    Calculates VIF each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param thresh: (default 5) the threshould VIF value. If the VIF of a variable is greater than thresh, it should be removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constatnt item to the data. add_constant is a function from statsmodels (see the import above)
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()

        # drop the const
        vif_df = vif_df.drop('const')
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        # If there are multiple variabels with VIF>thresh, only one of them is removed. This is because we want to keep as many variables as possible
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

```python
model_performance_expenditure = sm.OLS(endog=performance_expenditure_final[['score']], exog=sm.add_constant(df_predictors_selected_VIF)).fit()
model_performance_expenditure.summary()
```

```python
# Residuals vs. Fitted plot

# Set the figure size
plt.figure(figsize=(8, 6))  
# alpha = 0.5: controlling the transparency of the markers. Here alpha = 0.5 means that the markers are semi-transparent.
# s = 20: controlling the size of the markers. The default value is plt.rcParams['lines.markersize'] ** 2. Here we set it as 20.
# To get the default value of marker size, run the following line
# print(plt.rcParams['lines.markersize'] ** 2) 
plt.scatter(model_performance_expenditure.fittedvalues, model_performance_expenditure.resid, alpha = 0.5, s=20)
# plt.scatter(model_bike_rental.fittedvalues, model_bike_rental.resid, alpha = 0.5)
# adding title and labels
plt.xlabel('Fitted performance expenditure')
plt.ylabel('Residual')
plt.title('Residual vs. Fitted Plot of performance expenditure')
# Adding grid lines
plt.grid(True)
plt.show()
```
